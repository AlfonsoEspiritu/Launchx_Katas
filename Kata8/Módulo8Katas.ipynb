 1: Creación de diccionarios de Python

```python
Planeta={
    'nombre': 'Marte',
    'Lunas': 2

}
```

```python
print(f'{Planeta["nombre"]} tiene {Planeta["Lunas"]} Lunas')
```

```python
Planeta['Circunferencia (Km)'] = {
    'polar' : 6752,
    'acuatorial' : 6792
}
```

```python
print(f'{Planeta["nombre"]} tiene una circunferencia polar de {Planeta["Circunferencia (Km)"]["polar"]}')
```

2: Programación dinámica con diccionarios

```python
planet_moons = {
    'mercury': 0,
    'venus': 0,
    'earth': 1,
    'mars': 2,
    'jupiter': 79,
    'saturn': 82,
    'uranus': 27,
    'neptune': 14,
    'pluto': 5,
    'haumea': 2,
    'makemake': 1,
    'eris': 1
}
```

```python
moons = planet_moons.values()
planets= len(planet_moons.keys())
```

```python
total_moons= 0
for moon in moons:
    total_moons = total_moons + moon
```

```python
prom=total_moons/planets
print(prom)
```
