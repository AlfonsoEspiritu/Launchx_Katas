2 - Trabajando con datos de una lista

```python
planets = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno', 'Neptuno']
```

```python
nom_planet = input('Escribe el nombre del planeta "utilize una  letra mayuszula al comienzo del planeta"')
```

```python
planet = planets.index (nom_planet)
```

```python
print('Estos son los planetas mas cercanos: '+ str(nom_planet))
print(planets[0:planet])
```

```python
print('Estos son los planetas mas cercanos: '+ str(nom_planet))
print(planets[planet+1:])
```
