```python
Moon = ['average', 'temperature', 'distance ']
```

```python
for Sentence in temp.split():
    for Moon in Moon:
        if Moon in Sentence:
            print(Sentence)
            break
```

```python
for Sentence in temp.split():
    for Moon in Moon:
        if Moon in Sentence:
            print(Sentence.replace('C','Celcius'))
            break
```
