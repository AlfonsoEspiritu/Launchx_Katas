1 - Creaci√≥n de un bucle "while"

```python
new_planet = "Agregue un planeta"
planets = ['Mercurio','Venus','Marte','Jupiter','Saturno','Urano','Tierra']
```

```python
while new_planet.lower() != 'done':
    if new_planet:
        planets.append(new_planet)
    new_planet = input('Enter a new planet ')
    
```

2 - Creacion de un ciclo for

```python
for planet in planets:
    print(planet)
```
